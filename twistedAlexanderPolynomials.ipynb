```python
import sympy as sp

K = sp.symbols("K")
H = sp.symbols("H")
s = sp.symbols("s")
t = sp.symbols("t")
n1 = sp.symbols(r"\eta_1")
n2 = sp.symbols(r"\eta_2")
rho = sp.symbols(r"\rho")

n = 3
m = 1

for j in range (3, n+1, 1): # adjust n here
    polynomial = [1, t**(4*j+6)]
    
    for k in range (1, m+1, 1): # adjust m here
        alpha = (s**2-1)*s**(2*j)* \
                    (-k**6*(s-1)*s**2*(s**(2*j+1)+1)
                    +k**4*(s**(2*j+2)*(s**4-2*s**2+3*s-1) + s**4-3*s**3+2*s**2-1)
                    -k**2*s*(s**(2*j)*(2*s**3-s**2+1) - s*(s**3-s+2))
                    +s**2*(s**(2*j)-s**2))
        
        beta = k**7*s**(2*j+2)*(s**2-1)*(s**3+1) \
                    -k**5*s**3* \
                    (s**(4*j)*(s**3-s**2+1)
                    +s**(2*j-2)*(s-1)*(s**3+s+1)*(s**3+s**2+1)
                    -(s**3-s+1)) \
                    +k**3*s**2*(s**3+1)*(s**(2*j)-1)*(s**(2*j)+s**2) \
                    -k*s**3*(s**(2*j)-s**2)*(s**(2*j)+s)

        H = 1-k**2*s+k**2*s**(2*j+1) - s**(2*j-2)
        
        n1 = k*alpha - k*s**(2*j+1)*alpha + s**(2*j)*beta + k**2*s**(2*j)*beta
        
        n2 = -k*s*alpha + k*s**(2*j+1)*alpha - s**(2*j)*beta - s**(2*j+1)*beta
        
        for i in range (0, 2*3-1, 1):
            
            if i == 2*j - 1:
                lambda_i = (s**(j-1)-s**(-(j-1)))/(s-s**(-1)) - (s**2-1)*n1/(H*s**j*beta)
                polynomial.append(lambda_i * (t**(i+3) + t**(4*j-i+3)))
                
            elif 0 <= i and i <= 2*n-2 and i%2 == 0:
                lambda_i = (s*(k**2+1))/(k) * (s**(i/2) - (n1+n2)/(H*beta)*(s**(i/2+1)-s**(-i/2-1)))
                polynomial.append(lambda_i * (t**(i+3) + t**(4*j-i+3)))
                
            elif 0 <= i and i <= 2*n-2 and i%2 == 1:
                lambda_i = (s**((i-1)/2)-s**(-(i-1)/2))/(s-s**(-1))
                polynomial.append(lambda_i * (t**(i+3) + t**(4*j-i+3)))
           
        
        


polynomial = sp.expand(sp.simplify(sum(polynomial)))
polynomial
```
